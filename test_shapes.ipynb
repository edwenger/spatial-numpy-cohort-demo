```python
import os

from bokeh.models import GeoJSONDataSource
from bokeh.plotting import figure, show, output_notebook
import geopandas as gpd
```

```python

path = os.path.join("GRID3", "GRID3_NGA_-_Operational_LGA_Boundaries", "GRID3_NGA_-_Operational_LGA_Boundaries.shp")
lgas = gpd.read_file(path)
lgas["geometry"] = lgas["geometry"].to_crs(crs="EPSG:4326")

lgas = lgas[lgas.statename.isin([
    "Jigawa", "Kano", 
    # "Kaduna", "Katsina", "Zamfara", "Sokoto", "Kebbi", "Yobe", "Borno"
    ])]
```

```python
lgas.iloc[0]
```

```python
output_notebook()

geo_source = GeoJSONDataSource(geojson=lgas.to_json())

p = figure(width=800, tooltips=[('LGA', '@lganame'), ('State', '@statename')])
p.patches('xs', 'ys', source=geo_source, fill_alpha=0.4, fill_color="gray", line_color="black", line_width=0.5)

show(p)
```
